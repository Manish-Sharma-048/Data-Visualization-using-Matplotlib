Data Visualization : 
    Graphical representation of Data : 
        There are 2 ways of doing that. 
        1) coding = (matplotlib, seaborn, etc) using Python
        2) tool(software) = (Excel, Tableau, Power BI, etc) requires human involvement

Matplotlib lib supports 2 dimensional graphs (x,y axis). If we add some extra code, then we can create a 3D graph.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Scatter Graph 2D: The x,y points which we need to show can be stored in list, tuple or series. We need to make 2 lists 1 for x axis and 1 for y axis. In scatter graphs both y and x axis need to be mentioned.
 

```python
x = [150,145,120,160,154,180]
y = [60,55,70,75,90,100]
plt.figure(figsize=(10,6))
plt.scatter(x,y,marker='o',color='r',s=20)
plt.scatter(x,y)
plt.xticks(range(100,200,10))
plt.yticks(range(40,120,10))
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height v/s weight")
#plt.text(120,100,"Hello",size = 20,color='blue')
for xp,yp in zip(x,y):
    plt.text(xp,yp,f"({xp},{yp})",size = 10,color='black')
plt.grid(color = 'r', linestyle = '--',linewidth = .5)
#plt.show()
plt.savefig("ht_vs_wt.png")
```
